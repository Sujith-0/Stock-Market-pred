```python
import yfinance as yf
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
if os.path.exists("sensex.csv"):
    sensex = pd.read_csv("sensex.csv", index_col=0)
else:
    sensex = yf.Ticker("^BSESN").history(period="max")
    sensex.to_csv("sensex.csv")

sensex.index = pd.to_datetime(sensex.index)
```

```python
# Keep relevant features
sensex = sensex[["Open", "High", "Low", "Close", "Volume"]]
sensex = sensex.loc["2010-01-01":].copy()
```

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(sensex)

scaled_df = pd.DataFrame(scaled_data, index=sensex.index, columns=sensex.columns)
```

```python
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(sequence_length, len(data)):
        X.append(data[i-sequence_length:i])
        y.append(1 if data[i, 3] > data[i-1, 3] else 0)  # compare close prices
    return np.array(X), np.array(y)

sequence_length = 60
X, y = create_sequences(scaled_data, sequence_length)
```

```python
split = int(0.8 * len(X))
X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]
```

```python
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
predictions = model.predict(X_test)
predicted_classes = (predictions >= 0.5).astype(int)

plt.figure(figsize=(12,6))
plt.plot(predicted_classes, label="Predicted")
plt.plot(y_test, label="Actual", alpha=0.6)
plt.legend()
plt.title("LSTM - Actual vs Predicted")
plt.show()
```
